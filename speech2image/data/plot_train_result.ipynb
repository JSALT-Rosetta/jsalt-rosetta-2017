```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

tr_acc_sum = np.loadtxt('train_acry_sumall.txt')
tr_acc_rand = np.loadtxt('train_accuracy_scnn_rand_cost.txt')
niter = min(tr_acc_sum.shape[0], tr_acc_rand.shape[0])
plt.figure()
plt.plot(np.arange(niter), tr_acc_sum[0:niter], label='Sum of all cost')
plt.plot(np.arange(niter), tr_acc_rand[0:niter], label='Sum of rand cost')
plt.xlabel('number of train iteration')
plt.ylabel('train accuracy')
plt.title('Training Curve')
plt.legend(loc='best')
```
