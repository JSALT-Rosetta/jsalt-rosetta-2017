```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

tr = np.loadtxt('tr_scnn_40k_30iter.txt')
tx = np.loadtxt('dev_scnn_40k_30iter.txt')
niter = 30
nsteps = tr.shape[0]
iters = np.linspace(1, niter, nsteps)
plt.figure()
plt.plot(iters, tr, label='train')
plt.plot(iters, tx, label='test')
plt.xlabel('Number of iterations')
plt.ylabel('Accuracy')
plt.legend(loc='best')
```
